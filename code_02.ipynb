# 1a parte

No espaço de código abaixo, plote as 3 imagens, uma ao lado da outra. Ainda, plote o histograma de cada canal de cor (R - Vermelho, G - Verde e B - Azul) abaixo da respectiva imagem.

```
#printando imagens
imagem1 = Image.open('/content/drive/My Drive/Computação gráfica/bailey-zindel-NRQV-hBF10M-unsplash.jpg')
imagem2 = Image.open('/content/drive/My Drive/Computação gráfica/claudio-testa--SO3JtE3gZo-unsplash.jpg')
imagem3 = Image.open('/content/drive/My Drive/Computação gráfica/johannes-plenio-RwHv7LgeC7s-unsplash.jpg')

plt.figure(figsize=(20,4))

#primeira
plt.subplot(131)
plt.title('Imagem 1')
plt.axis('off')
plt.imshow(imagem1)
#segunda
plt.subplot(132)
plt.title('Imagem 2')
plt.axis('off')
plt.imshow(imagem2)
#terceira
plt.subplot(133)
plt.title('Imagem 3')
plt.axis('off')
plt.imshow(imagem3)

plt.show()

#gerar arrays das imagens
img1_arr = np.array(imagem1)
img2_arr = np.array(imagem2)
img3_arr = np.array(imagem3)

#gerar histograma da imagem 1

red = img1_arr[:,:,0]
green = img1_arr[:,:,1]
blue = img1_arr[:,:,2]
plt.figure(figsize=(20,4))

plt.subplot(131)
plt.title("Imagem1 Red")
plt.hist(red.ravel(), color='red')

plt.subplot(132)
plt.title("Imagem1 Green")
plt.hist(green.ravel(), color='green')

plt.subplot(133)
plt.title("Imagem1 Blue")
plt.hist(blue.ravel(), color='blue')

#gerar histograma da imagem 2

red = img2_arr[:,:,0]
green = img2_arr[:,:,1]
blue = img2_arr[:,:,2]
plt.figure(figsize=(20,4))

plt.subplot(131)
plt.title("Imagem2 Red")
plt.hist(red.ravel(), color='red')

plt.subplot(132)
plt.title("Imagem2 Green")
plt.hist(green.ravel(), color='green')

plt.subplot(133)
plt.title("Imagem2 Blue")
plt.hist(blue.ravel(), color='blue')

#gerar histograma da imagem 2

red = img3_arr[:,:,0]
green = img3_arr[:,:,1]
blue = img3_arr[:,:,2]
plt.figure(figsize=(20,4))

plt.subplot(131)
plt.title("Imagem3 Red")
plt.hist(red.ravel(), color='red')

plt.subplot(132)
plt.title("Imagem3 Green")
plt.hist(green.ravel(), color='green')

plt.subplot(133)
plt.title("Imagem3 Blue")
plt.hist(blue.ravel(), color='blue')

```

Olhando para os gráficos gerados, discorram comparando os resultados.

Podemos perceber que a imagem 2 e 3 apresentam os niveis mais altos de vermelho, enquanto a imagem 1 apresenta um nivel menor de vermelho, ficando mais concentrado enter 0 e 50. Nos tons de verde, as imagens 2 e tres também apresentam os maiores niveis entre 50 e 200, enquanto a imagem 1 apresenta mais verde com valores acima de 200. No azul, as imagens 1 e 2 possuem valores mais expressivos, enquanto a terceira imagem apresenta apenas valores maiores próximos do 0.
